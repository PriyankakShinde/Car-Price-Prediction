# Model Building

```python
from sklearn.model_selection import train_test_split
```

```python
df_train,df_test = train_test_split(df,train_size=0.7,test_size=0.3,random_state=100)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df.info()
```

```python
#we'll not scale every column which is normal
varlist=['symboling', 'doornumber', 'wheelbase', 'carlength', 'carwidth',
       'carheight', 'curbweight', 'cylindernumber', 'enginesize', 'boreratio',
       'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg',
       'highwaympg']
df_train[varlist]=scaler.fit_transform(df_train[varlist])
```

```python
df_train.head()
```

```python
y_train=df_train.pop('price')
```

```python
y_train
```

```python
df_train.columns
```

```python
X_train = df_train
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
#for linear reg when e have 2 values formulae becomes y=a1x1+a2x2+b
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
#recursive feature elimination RFE so we can get to kno how each column is performing
from sklearn.feature_selection import RFE
rfe1=RFE(lm,n_features_to_select=15) #get top 15 columns set by set
rfe1.fit(X_train,y_train)
```

```python
rfe1.ranking_
```

```python
rfe1.support_
```
